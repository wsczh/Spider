```python
import requests
import json
import re
import time
import pandas as pd
pd.set_option('display.max_rows',1000)
```

```python
headers = {
    "authorization": "",
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"
}
```

```python
ses = requests.session()
res = ses.get('https://api.cc98.org/Topic/5015041/post?from=0&size=10', headers = headers)
```

```python
cnt = 0
dic = {}
last_url = 'https://api.cc98.org/Topic/5015041/post?from=0&size=10'
while last_url and cnt <12:
    time.sleep(3)
    cnt += 1
    res = ses.get(last_url)
    res = json.loads(res.text)[0]['content']
    res = res.replace(f'[/align]', '').split('[size=4]')
    last_items = re.search('\[url=(.*)\](.*)\[/url\]', res[0])
    
    if last_items:
        last_url = (last_items.group(1)+'/post?from=0&size=10').replace('www','api')
        last_title = last_items.group(2).strip()
    else:
        last_url = None
    
    for topic in res[1:]:
        items = topic.split('\n')
        topic_name = items[0].replace('[b]','')
        if topic_name not in dic.keys():
            dic[topic_name] = []
        results = re.findall("\[url=(.*?)\](.*?)\[/url\]", topic) # ?最短匹配
        dic[topic_name].extend(results)
```

```python
dic.keys()
```

```python
items = []
for key, value in dic.items():
    for item in value:
        items.append([key, item[0], item[1]])
```

```python
items
```

```python
df = pd.DataFrame(items, columns = ['板块', '链接', '标题'])
df.drop(df[df['标题']=='推荐阅读投稿楼'].index, axis = 0, inplace = True) # 没用的链接
df['链接'] = df['链接'].apply(lambda x: 'https://www.cc98.org'+x if 'www' not in x else x) #补全链接
```
